```python
import pandas as pd
df=pd.read_csv(r"D:\HR_Analysis\HR_Analytics.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', None)
df.head()
```

```python
df.describe()
```

```python
# Check for missing values
missing = df.isnull().sum()
missing = missing[missing > 0]
print('Columns with missing values:')
print(missing)
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f'Number of duplicate rows: {duplicates}')
```

```python
print(f'Rows before removing duplicates: {len(df)}')
df = df.drop_duplicates().reset_index(drop=True)
print(f'Rows after removing duplicates: {len(df)}')
```

```python
# Confirm before dropping
cols_to_check = ['EmployeeCount', 'Over18', 'StandardHours']
for col in cols_to_check:
    print(f'{col} unique values: {df[col].unique()}')
```

```python
print(f'Columns before dropping: {df.shape[1]}')
columns_to_drop = ['EmployeeCount', 'Over18', 'StandardHours']
df = df.drop(columns=columns_to_drop)

print(f'Columns after dropping: {df.shape[1]}')
print('Remaining columns:', list(df.columns))
```

```python
# See median by JobRole before filling
print('Median YearsWithCurrManager by JobRole:')
print(df.groupby('JobRole')['YearsWithCurrManager'].median())
```

```python
# Fill nulls with JobRole group median
df['YearsWithCurrManager'] = df.groupby('JobRole')['YearsWithCurrManager'].transform(
    lambda x: x.fillna(x.median())
)

# Confirm no nulls remain
print(f'Missing values in YearsWithCurrManager after fix: {df["YearsWithCurrManager"].isnull().sum()}')

# Convert to int since years should be whole numbers
df['YearsWithCurrManager'] = df['YearsWithCurrManager'].astype(int)
```

```python
df['AttritionFlag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

print('AttritionFlag value counts:')
print(df['AttritionFlag'].value_counts())
print(f'\nOverall Attrition Rate: {df["AttritionFlag"].mean()*100:.1f}%')
```

```python
# Final null check — should be zero across all columns
print('Null values per column (should all be 0):')
print(df.isnull().sum()[df.isnull().sum() > 0])

if df.isnull().sum().sum() == 0:
    print('\nNo missing values remain.')
```

```python
# Sanity check on key columns — ranges should make sense
sanity_cols = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'JobSatisfaction', 'WorkLifeBalance']
df[sanity_cols].describe()
```

```python
# Final shape and column list
print(f'Final dataset shape: {df.shape}')
print(f'\nFinal columns ({df.shape[1]}):')
for col in df.columns:
    print(f'  - {col} ({df[col].dtype})')
```

```python
df.to_csv('hr_analytics_cleaned.csv', index=False)

print('✓ Cleaned dataset saved as: hr_analytics_cleaned.csv')
print(f'  Rows: {df.shape[0]}')
print(f'  Columns: {df.shape[1]}')
```
